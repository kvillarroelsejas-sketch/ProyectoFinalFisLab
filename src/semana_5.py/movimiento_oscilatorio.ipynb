```python
import numpy as np
import matplotlib.pyplot as plt


k = 25.0  # ejemplo [N/m]

m_base = 0.5  #ejemplo[kg]
M = 9 * m_base  # La masa para esta gr√°fica es 9m
# resultado => M = 4.5[kg]

# amplitud (A)
A = 0.10  # [m] (Ejemplo: 10 cm)

# t=0, p=0
phi = 0.0  # [rad]

# frecuencia angular
# w = sqrt(k / M)
omega = np.sqrt(k / M)

T = 2 * np.pi / omega
print(f"La frecuencia angular (w) es: {omega:.2f} rad/s")
print(f"El periodo de oscilaci√≥n (T) es: {T:.2f} segundos")

# definir el rango de (t)
t_max = 4 * T
puntos = 500
t = np.linspace(0, t_max, puntos)

# x(t) = A * cos(w*t + phi)
x = A * np.cos(omega * t + phi)

# graficar los resultados
plt.figure(figsize=(10, 6))

# gr√°fica de la curva de posici√≥n vs tiempo
plt.plot(t, x, label=f'Masa = {M} kg', color='purple')

# amplitud m√°xima y m√≠nima
plt.axhline(A, color='green', linestyle='--', alpha=0.6, label='Amplitud +A')
plt.axhline(-A, color='green', linestyle='--', alpha=0.6, label='Amplitud -A')
plt.axhline(0, color='black', linestyle='-', linewidth=0.5, label='Equilibrio')

plt.title(f'Movimiento Arm√≥nico Simple (MAS) para Masa $9m$ ($M={M}$ kg)', fontsize=14)
plt.xlabel('Tiempo $t$ (s)', fontsize=12)
plt.ylabel('Posici√≥n $x$ (m)', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```
