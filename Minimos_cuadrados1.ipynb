```python
import numpy as np
import matplotlib.pyplot as plt

# Función para ajustar y graficar
def ajustar_y_graficar(t, x, masa_label):
    # Ajuste lineal
    A, B = np.polyfit(t, x, 1)
    x_fit = A * np.array(t) + B
    error = np.std(np.array(x) - x_fit)

    # Gráfica
    plt.figure()
    plt.plot(t, x, 'o', label='Datos experimentales')
    plt.plot(t, x_fit, 'r-', label=f'Ajuste: x = {A:.2f}t + {B:.2f}')
    plt.title(f'Masa {masa_label} - Error: {error:.4f}')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Posición (m)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Datos para cada masa
datos = {
    'm':   [ [0, 1, 2, 3, 4], [0.0, 0.8, 0.0, -0.8, 0.0] ],
    '2m':  [ [0, 1, 2, 3, 4], [0.0, 0.6, 0.0, -0.6, 0.0] ],
    '3m':  [ [0, 1, 2, 3, 4], [0.0, 0.5, 0.0, -0.5, 0.0] ],
    '4m':  [ [0, 1, 2, 3, 4], [0.0, 0.4, 0.0, -0.4, 0.0] ],
    '5m':  [ [0, 1, 2, 3, 4], [0.0, 0.35, 0.0, -0.35, 0.0] ],
    '6m':  [ [0, 1, 2, 3, 4], [0.0, 0.3, 0.0, -0.3, 0.0] ]
}

# Ejecutar para cada masa
for masa, (t, x) in datos.items():
    ajustar_y_graficar(t, x, masa)
```
